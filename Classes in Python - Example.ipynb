```python
class Student:
    def __init__(self, name):
        self.name = name
        self.grades = []

    def add_grade(self, new_grade):
        self.grades.append(new_grade)

    def get_average(self):
        if len(self.grades) == 0:
            return 0
        return sum(self.grades) / len(self.grades)
    def grade_report(self):
        print(f"Grade report for {self.name}:")
        for i, grade in enumerate(self.grades, start=1):
            if grade >= 90:
                letter = 'A'
            elif grade >= 80:
                letter = 'B'
            elif grade >= 70:
                letter = 'C'
            elif grade >= 60:
                letter = 'D'
            else:
                letter = 'F'
            print(f"  Assignment {i}: {grade} ({letter})")

    def has_passed(self):
        return all(grade >= 40 for grade in self.grades)

        
```

```python
Student1 = Student("Alice")
Student1.add_grade(85)
Student1.add_grade(92)
Student1.add_grade(78)
print(Student1.get_average())
Student1.grade_report()
Student1.has_passed()
```
